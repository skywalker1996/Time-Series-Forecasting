```python
from dataset import *
from torch.utils.data import Dataset, DataLoader 
from matplotlib import pyplot as plt


train_df = pandas.read_csv('./dataset/server.csv')
raw_data = train_df.values[:,1]
```

```python
processed_data = [raw_data[0]]
EMA = raw_data[0]
alpha = 0.05
for rtt in raw_data:
    EMA = alpha * rtt + (1-alpha)*EMA
    processed_data.append(EMA)
```

```python
plt.figure(figsize=(20,10))#
plt.title('Examples of line chart',fontsize=20)#
plt.xlabel('time',fontsize=14)#x
plt.ylabel('elec',fontsize=14)#y


plt.plot(list(range(5000)),processed_data[:5000], label='RTT')
# plt.plot(list(range(300)),labels_cla[300:600]*50+80, label='pridiction')
plt.legend(loc='upper left')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

# class EMA
server_sending_rate = 10

def rising_sequence(start, end):
	return list(np.arange(start,end,2))

def falling_sequence(start, end):
	return list(np.arange(start,end,-2))
```

```python
Bandwidth_seq = []

## 1.2 * server_sending_rate
Bandwidth_seq+=[server_sending_rate*1.2+np.random.randn() for i in range(50)]
## rising
Bandwidth_seq+=rising_sequence(server_sending_rate*1.2,server_sending_rate*2)

## 2 * server_sending_rate
Bandwidth_seq+=([server_sending_rate*2+np.random.randn() for i in range(50)])

## falling
Bandwidth_seq+=(falling_sequence(server_sending_rate*2,server_sending_rate*1.2))

## 1.2 * server_sending_rate
Bandwidth_seq+=([server_sending_rate*1.2+np.random.randn() for i in range(50)])

## falling
Bandwidth_seq+=(falling_sequence(server_sending_rate*1.2,server_sending_rate*0.5))

## 0.5 * server_sending_rate
Bandwidth_seq+=([server_sending_rate*0.5+np.random.randn() for i in range(50)])

## rising
Bandwidth_seq+=(rising_sequence(server_sending_rate*0.5,server_sending_rate*1.2))

f = open('../traces/trace_5.txt', 'w+')
for bd in Bandwidth_seq:
	f.write(str(round(bd,2))+'\n')

f.close()

plt.figure(figsize=(15,10))
plt.title("demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(list(range(len(Bandwidth_seq))),Bandwidth_seq) 
plt.show()
```
